```python
years = [2021,2022,2023]
vals = []
for i in years:
    low = yearly_rolling.loc[yearly_rolling[i] >25].head(1).index[0]
    high = yearly_rolling.loc[yearly_rolling[i] >75].head(1).index[0]
    mid = yearly_rolling.loc[yearly_rolling[i] >50].head(1).index[0]
    vals.append([i,low,mid,high])
yearly_breakdown:pd.DataFrame = pd.DataFrame(vals,columns=["Year","25%","50%", "75%"])
yearly_breakdown = yearly_breakdown.T
yearly_breakdown.rename(columns=yearly_breakdown.iloc[0], inplace=True)
yearly_breakdown
#yearly_breakdown.drop("Year")
```

```python
common_columns = list(set(productivity.columns).intersection(yearly_rolling.columns))

common_columns
```

```python
dframe = pd.DataFrame()


v = None
for i in map["Region"]:
    print(i)
    valid = data['Provider Org Name'].isin(map.loc[map.Region==i].Hospital)

    valid = data.loc[valid]
    if valid.empty:continue
    #result_df.columns = pd.to_datetime(result_df.columns, format='%b-%y').year

    valid = valid.iloc[:,4:].groupby(valid.Period).mean().T.cumsum()#.iloc[:,4:]#.groupby().mean().T.cumsum()
    valid.columns = pd.to_datetime(valid.columns, format="%B-%Y").year
    valid = valid.T
    valid = valid.groupby(valid.index).mean().T
    vals = []
    v = valid
    for x in common_columns:
        #print(x)
        value = productivity.loc[productivity["Region name"] == i][x].values[0]

        val = valid.loc[valid[x] > percentile].head(1)[x]
        high = val.values[0]
        low = valid.iloc[int(val.index[0])-2][x]
        actual_index = float(val.index[0])+ ((percentile - low) / (high - low))
        print(value,actual_index)
        vals.append([i,value,actual_index,x])
    dframe:pd.DataFrame = pd.concat((dframe,pd.DataFrame(vals,columns=["Region","Productivity", "Wait Time","Year"])),ignore_index=True)
dframe
```
