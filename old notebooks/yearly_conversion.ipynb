```python
import pandas as pd
import plotly
df = pd.read_csv("data/all_hospital.csv")
map = pd.read_excel("data/mapping.xlsx")
```

```python
region = "Peterborough"
location = "BUCKINGHAMSHIRE HEALTHCARE NHS TRUST"
```

```python
percentage = df.columns[14:-1]
percentage_new = df
percentage_new[percentage] = df[percentage].div(df.Total, axis=0)*100
```

```python
result_df:pd.DataFrame = percentage_new.loc[df['Provider Org Name'].isin(map.loc[map.Region==region].Hospital)].T#.iloc[14:-1]
result_df.columns = result_df.iloc[0]
result_df:pd.DataFrame = result_df.iloc[1:].T

result_df = result_df.T.iloc[:-1]
result_df
```

```python
result_df.median()
```


# waiting time per month in selected region

```python

mean_ar=result_df.groupby(result_df.columns, axis=1).mean()
accumilative = mean_ar.cumsum()
plotly.plot(accumilative,"line")
```

```python
yearly_rolling = result_df.copy(deep=True)
#yearly_rolling.columns = pd.to_datetime(yearly_rolling.columns, format='%B-%Y').year
#yearly_rolling = yearly_rolling.T.groupby(["Period"]).mean().T
yearly_rolling
```

# breakdown for the region per year

```python
plotly.plot(yearly_rolling, "line")
```

# accumilative of this data

```python
accumilative_year_rolling = yearly_rolling.cumsum()
plotly.plot(accumilative_year_rolling, "line")
```

# index for each percentile

```python
percentile = 50
year = 2021
accumilative_year_rolling.loc[accumilative_year_rolling[year] > percentile][2021].head(1).index
```

```python

years = [2021,2022,2023]
vals = []
for i in years:
    low = accumilative_year_rolling.loc[accumilative_year_rolling[i] >25].head(1).index[0]
    high = accumilative_year_rolling.loc[accumilative_year_rolling[i] >75].head(1).index[0]
    mid = accumilative_year_rolling.loc[accumilative_year_rolling[i] >50].head(1).index[0]
    vals.append([i,low,mid,high])
yearly_breakdown:pd.DataFrame = pd.DataFrame(vals,columns=["Year","25%","50%", "75%"])
yearly_breakdown = yearly_breakdown.T
yearly_breakdown.rename(columns=yearly_breakdown.iloc[0], inplace=True)
yearly_breakdown.drop("Year")
```

```python
percentile = 25
```

```python
data = pd.ExcelFile("data/productivity.xlsx")

y = pd.read_excel(data, "A1", skiprows=[1,2,3])
#y = y.drop(0)
y.rename(columns=y.iloc[0], inplace = True)
y = y.drop(0)
for column in y.columns:
    if isinstance(column, float):
        y.rename(columns={column: int(column)}, inplace=True)

y.loc[y["Region name"] == region]
```

```python
# find common data
common_columns = list(set(y.columns).intersection(yearly_breakdown.columns))
common_columns
```

```python
dframe = pd.DataFrame(columns=["Region","Productivity", "Wait Time","Year"])
for i in y["Region name"]:
    result_df:pd.DataFrame = percentage_new.loc[df['Provider Org Name'].isin(map.loc[map.Region==i].Hospital)].T#.iloc[14:-1]
    if result_df.empty:continue
    result_df.columns = result_df.iloc[0]
    result_df:pd.DataFrame = result_df.iloc[1:].T
    result_df = result_df.drop(columns=["Provider Parent Org Code", "Provider Parent Name", "Provider Org Code", "Commissioner Parent Org Code" ,"Commissioner Parent Name", "Commissioner Org Name", "RTT Part Description", "Treatment Function Code", "Treatment Function Name", "RTT Part Type", "Commissioner Org Code"]).T.iloc[2:-1]
    result_df.columns = pd.to_datetime(result_df.columns, format='%b-%y').year
    result_df = result_df.T.groupby(["Period"]).mean().T.cumsum()
    vals = []
    for x in common_columns:
        
        
        value = y.loc[y["Region name"] == i].head(1)[x].values[0]
        val = result_df.loc[result_df[x] > percentile].head(1)[x]
        
        high = val.values[0]
        low = y.loc[float(val.index[0])-1][x]
        actual_index = float(val.index[0])+ ((percentile - low) / (high - low))
        
        vals.append([i,value,actual_index,x])
    dframe:pd.DataFrame = pd.concat((dframe,pd.DataFrame(vals,columns=["Region","Productivity", "Wait Time","Year"])),ignore_index=True)
dframe
    
```
